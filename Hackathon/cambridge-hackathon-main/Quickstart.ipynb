# Agents (Chatbot)

Run the example below to create an agent that will greet you and initiate a voice conversation. 
Speak back to it to initiate a response.

1. Speak into the microphone, and the system will transcribe your speech in real time.
2. The transcribed text is sent to the LLM to generate a response.
3. The response is converted to speech using the TTS module and played back to you.
4. Repeat the process to continue the conversation.

```python
client = Neuphonic(api_key=api_key)

neurodivergence = "ADHD"

# greeter
agent_id = client.agents.create(
    name='Agent 1',
    prompt=f"You're a chatbot for a neurodivergence mentoring app. The user has specified they have {neurodivergence}. Providing short (maximum 3 sentence) responses, find out what their immediate concern is, then take me through one actionable step at a time, waiting for my responses between each one.",
    greeting='Hi, how can I help you today?'
).data['id']

# Emotional coaching agent
emotion_coach = client.agents.create(
    name='Agent 1',
    prompt=f"You're a chatbot for a neurodivergence mentoring app. The user has specified they have {neurodivergence}. Providing short (maximum 3 sentence) responses, find out what their immediate concern is, then take me through one actionable step at a time, waiting for my responses between each one.",
    greeting='Hi, how can I help you today?'
).data['id']

# task overwhelm agent
agent_id = client.agents.create(
    name='Agent 1',
    prompt=f"You're a chatbot for a neurodivergence mentoring app. The user has specified they have {neurodivergence}. Providing short (maximum 3 sentence) responses, find out what their immediate concern is, then take me through one actionable step at a time, waiting for my responses between each one.",
    greeting='Hi, how can I help you today?'
).data['id']

# agent_id = client.agents.create(
#     name='Agent 1',
#     prompt="I want you to pretend you're a chatbot for a neurodivergence mentoring app, and provide short (maximum 3 sentence) responses accordingly. Assume I have ADHD. First, ask what I need help with. Then, give me one immediately actionable step at a time, waiting  for my responses.",
#     greeting='Hi, how can I help you today?'
# ).data['id']

agent = Agent(client, agent_id=agent_id, tts_model='neu_hq')

await agent.start()
```
